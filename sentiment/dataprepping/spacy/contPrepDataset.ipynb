```python
import json
import spacy

with open('../datasets/3_text_and_gender.json') as json_file:
    data = json.load(json_file)

nlp = spacy.load("en_core_web_lg")
```

```python
from collections import OrderedDict
import json

with open('../datasets/4_word_weight_m_SM.json') as json_file:
    m_data = json.load(json_file)

with open('../datasets/4_word_weight_w_SM.json') as json_file:
    w_data = json.load(json_file)

m_descending = OrderedDict(sorted(m_data.items(), key=lambda kv: kv[1], reverse=True))
w_descending = OrderedDict(sorted(w_data.items(), key=lambda kv: kv[1], reverse=True))

with open('../datasets/4_word_weight_m_SM_ordered.json', "w") as outFile:
    json.dump(m_descending, outFile)

with open('../datasets/4_word_weight_w_SM_ordered.json', "w") as outFile:
    json.dump(w_descending, outFile)
```

```python
doc = nlp(data['articles'][0]['text'])

assert doc.has_annotation("SENT_START")
for sent in doc.sents:
    print(sent.text)
    for token in sent:
        print(token.text)
        

# for token in doc:
#     print(token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
#             token.shape_, token.is_alpha, token.is_stop)
```
